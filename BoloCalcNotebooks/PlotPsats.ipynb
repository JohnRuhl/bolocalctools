```python
NETcmb1 = unpack.vary_outputs['V0']['Tel']['Cam']['Summary']['Cam_1']['Detector NET_CMB'][:,0]
NETcmb2 = unpack.vary_outputs['V0']['Tel']['Cam']['Summary']['Cam_2']['Detector NET_CMB'][:,0]
Popt1 = unpack.vary_outputs['V0']['Tel']['Cam']['Summary']['Cam_1']['Optical Power'][0,0]
Popt2 = unpack.vary_outputs['V0']['Tel']['Cam']['Summary']['Cam_2']['Optical Power'][0,0]
```

```python
Popt2
```

```python
plt.clf()
plt.plot(psat1,NETcmb1,'.-b',label='90GHz')
plt.plot(psat2,NETcmb2,'.-r',label='150GHz')
plt.plot([Popt1,Popt1],[0,600],'b')
plt.text(Popt1,200,'Popt_90GHz',color='blue')
plt.plot([Popt2,Popt2],[0,600],'r')
plt.text(Popt2,200,'Popt_150GHz',color='red')
plt.xlabel('Psat [pW]')
plt.ylabel('Detector NET_CMB [uKrtsec]')
plt.xlim([0,20])
plt.ylim([0,600])
plt.legend()
plt.grid()
plt.title('Example Experiment')
```
